## Guessing Game

```python
import random
random_number=random.randint(1,100)
guessed_number=False
while not guessed_number:
    try:
        user_guess=int(input('Enter your guessing:'))
        if user_guess>random_number:
            print('Too high')
        elif user_guess<random_number:
            print('Too low')
        else:
            print('Congrats, You\'ve guessed the correct number')
            guessed_number=True
    except ValueError:
        print('Enter a valid numerical value between 1 and 100')
```

## Number Guessor

```python
import random
X,Y=map(int,input("Enter the range in the format of X,Y where X is the start value and Y is the end value:").split(','))
random_number=random.randint(X,Y)
guessed_number=False
counter=0
while not guessed_number:
    try:
        user_guess=int(input('Enter your guessing:'))
        counter+=1
        if user_guess>random_number:
            print('Too high')
        elif user_guess<random_number:
            print('Too low')
        else:
            print('Congrats, You\'ve guessed the correct number')
            guessed_number=True
    except ValueError:
        print('Enter a valid numerical value between 1 and 100')
print(f'You took {counter} attempts to guess the correct number')
```

## Password Strength Checker

```python
import re
def password_checker(password):
    strength=0
    if len(password)>=8:
        strength+=1
    else:
        print('Your password should have atleast 8 characters minimum')
    if re.search(r'[A-Z]',password) and re.search(r'[a-z]',password):
        strength+=1
    else:
        print('Your password must be a combination of both Uppercase and Lowercase letters')
    if re.search(r'\d',password):
        strength+=1
    else:
        print('Your password must contain atleast one number')
    if re.search(r'[!@#$%^&*_]',password):
        strength+=1
    else:
        print('Your password must contain atleast one special character')
    if strength>3:
        print('Your password is very strong')
    elif strength>2:
        print('Your password is moderate')
    elif strength>1:
        print('Your password is too weak')
password=input("Enter the password:")
password_checker(password)
```

## Fibonacci Sequence

```python
def fibonacci_sequence(n):
    fib_seq = [0, 1]
    if n <= 0:
        print("Please enter a positive integer.")
    elif n == 1:
        print("Fibonacci sequence up to 1 term: [0]")
    else:
        for i in range(2, n):
            next_term = fib_seq[-1] + fib_seq[-2]
            fib_seq.append(next_term)
        print(f"Fibonacci sequence up to {n} terms: {fib_seq[:n]}")
num_terms = int(input("Enter the number of terms: "))
fibonacci_sequence(num_terms)
```

## File Manipulation

```python
import string

def count_words_in_file(file_path):
    word_count = {}
    try:
        with open(file_path, 'r') as file:
            for line in file:
                line = line.translate(str.maketrans("", "", string.punctuation)).lower()
                words = line.split()
                for word in words:
                    if word in word_count:
                        word_count[word] += 1
                    else:
                        word_count[word] = 1
        sorted_word_count = dict(sorted(word_count.items()))
        print("Word count in alphabetical order:")
        for word, count in sorted_word_count.items():
            print(f"{word}: {count}")
    except FileNotFoundError:
        print("The file was not found. Please check the file path.")
file_path = input("Enter the path to the text file: ")
count_words_in_file(file_path)
```
